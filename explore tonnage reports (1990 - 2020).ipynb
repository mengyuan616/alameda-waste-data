```python
import pandas as pd
```

```python
data = pd.DataFrame()
```

```python
for year in range(1990, 2021): 
#     if year == 2019：
    report = pd.read_excel(f'raw_data/landfill_sum/LandfillSummaryTonnage - { year }.xlsx', skiprows=4)
    del report['Unnamed: 1']
    
    #  report.rename(columns = {f'{year} Total'} : 'Total', inplace=True) # alternate way of changing column name
    report.columns = ['county', '1st Quarter','2nd Quarter','3rd Quarter','4th Quarter', 'Total']
    report['Year'] = year
    
    # delete last row
    report.drop(report.tail(1).index, inplace = True)
    
    data = data.append(report, ignore_index=True)
```
